```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.decomposition import PCA

pca_data = raw_data.drop(columns="Class", axis=1)
plot_column = raw_data["Class"]
```

```python
features_standard = (pca_data - pca_data.mean())/pca_data.std()
features_standard
```

```python
pca = PCA(n_components=2)
pca_scaled = pca.fit_transform(features_standard)
pca_scaled
```

```python
plot_data = pd.DataFrame(data=pca_scaled, columns=["pca1","pca2"])
plot_data["Class"] = plot_column

plot_data
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(data=plot_data, x='pca1', y='pca2', hue='Class', style='Class', palette='deep')

plt.title('PCA: Pumpkin Seed Classes in 2D Space')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Class', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(12, 6))

# Separate the dataset by class and plot histograms
for class_name, group in raw_data.groupby('Class'):
    # Plotting with density=True for normalized histograms
    plt.hist(group['Area'], bins=30, alpha=0.5, label=class_name, density=True)

plt.title('Normalized Histogram of Pumpkin Seed Area by Class')
plt.xlabel('Area')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
X = raw_data.drop("Class", axis=1)
y = raw_data["Class"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=47)

#using SVC model
scaler = StandardScaler()
model = SVC()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

clf = SVC(kernel='rbf', random_state=42)  
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')

accuracy_percent=round(accuracy,2)
print("Accuracy is:",accuracy_percent*100,"%")

print(classification_report(y_test, y_pred))
```
