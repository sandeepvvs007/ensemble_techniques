Movie Ratings (MPAA)
G (General Audiences): Suitable for all ages. Contains nothing that would offend parents if viewed by children.
PG (Parental Guidance Suggested): Some material may not be suitable for children. Parents are urged to give "parental guidance." May contain some material parents might not like for their young children.
PG-13 (Parents Strongly Cautioned): Some material may be inappropriate for children under 13. Parents are urged to be cautious. Some material may be inappropriate for pre-teenagers.
R (Restricted): Under 17 requires accompanying parent or adult guardian. Contains some adult material. Parents are urged to learn more about the film before taking their young children with them.
NC-17 (No One 17 and Under Admitted): Clearly adult. Children are not admitted.
TV Ratings (TV Parental Guidelines)
TV-Y (All Children): This program is designed to be appropriate for all children.
TV-Y7 (Directed to Older Children): This program is most appropriate for children age 7 and up.
TV-Y7-FV (Fantasy Violence): Programs with this rating may include more intense fantasy violence.
TV-G (General Audience): Most parents would find this program suitable for all ages. It contains little or no violence, no strong language, and little or no sexual dialogue or situations.
TV-PG (Parental Guidance Suggested): This program contains material that parents may find unsuitable for younger children. Parental guidance is advised.
TV-14 (Parents Strongly Cautioned): This program contains some material that many parents would find unsuitable for children under 14 years of age.
TV-MA (Mature Audience Only): This program is specifically designed to be viewed by adults and therefore may be unsuitable for children under 17.
Special Categories
NR (Not Rated): The film or show has not been submitted for a rating or is an uncut version of a film that was submitted.
UR (Unrated): This is similar to NR. It usually means that the film was released without a formal rating from the MPAA or is an uncut version of a film.
74 min, 84 min, 66 min: These are durations of the content, indicating the total runtime in minutes.
NaN (Not a Number): This is not a rating but a placeholder typically used in databases or spreadsheets to indicate missing or undefined values.
Each rating provides guidance to help viewers make informed choices about what movies or TV shows are appropriate for themselves or their children, based on the content's theme, language, violence, sex, and other sensitive elements.

```python
#unnesting the directors column, i.e- creating separate lines for each director in a movie
constraint1=data['director'].apply(lambda x: str(x).split(', ')).tolist()
df_new1 = pd.DataFrame(constraint1, index=data["title"])
df_new1 = df_new1.stack()
df_new1 = pd.DataFrame(df_new1.reset_index())
df_new1.rename(columns={0:'director'},inplace=True)
df_new1.drop(['level_1'],axis=1,inplace=True)
df_new1.head()
```

```python
constraint2 = data['cast'].apply(lambda x : str(x).split(', ')).to_list()
df_new2 = pd.DataFrame(constraint2, index=data['title'])
df_new2 = df_new2.stack()
df_new2=df_new2.reset_index()
df_new2.rename(columns={0 :'actor'}, inplace=True)
df_new2.drop(columns='level_1', inplace=True, axis=1)
df_new2
```

```python
constraint3 = data['country'].apply(lambda x : str(x).split(', ')).to_list()
df_new3 = pd.DataFrame(constraint3, index=data['title'])
df_new3 = df_new3.stack()
df_new3= df_new3.reset_index()
df_new3.rename(columns={0:'country'}, inplace=True)
df_new3.drop(columns='level_1', inplace=True, axis=1)
df_new3
```

```python
constraint4 = data["listed_in"].apply(lambda x : str(x).split(', ')).to_list()
df_new4 = pd.DataFrame(constraint4, index=data["title"])
df_new4 = df_new4.stack()
df_new4 = df_new4.reset_index()
df_new4.rename(columns={0:"listed_in"}, inplace=True)
df_new4.drop(columns="level_1",inplace=True, axis=1)
df_new4
```

```python
import numpy as np
df_new5= df_new1.merge(df_new2, on="title", how="inner").merge(df_new3, on="title", how="inner").merge(df_new4, on="title", how="inner")
df_new5
df_new5['actor'].replace(['nan'],['Unknown Actor'],inplace=True)
df_new5['director'].replace(['nan'],['Unknown Director'],inplace=True)
df_new5['country'].replace(['nan'],[np.nan],inplace=True)
df_new5
```

```python
df_final=df_new5.merge(data[['show_id', 'type', 'title', 'date_added',
       'release_year', 'rating', 'duration']],on=['title'],how='left')
df_final.head()
```
